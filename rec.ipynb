```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from PIL import Image 
```

```python
def is_point_in_rectangle(p, rect):
    """
    Check if a point lies within a rectangle.
    
    Args:
    p: A tuple (x, y) representing the point to test.
    rect: A list of tuples [(x1, y1), (x2, y2), (x3, y3), (x4, y4)] representing
          the rectangle vertices in clockwise or counterclockwise order.
    
    Returns:
    True if the point lies inside the rectangle, False otherwise.
    """
    def vector(a, b):
        """Compute the vector from point a to point b."""
        return (b[0] - a[0], b[1] - a[1])
    
    def dot_product(v1, v2):
        """Compute the dot product of two vectors."""
        return v1[0] * v2[0] + v1[1] * v2[1]
    
    def magnitude_squared(v):
        """Compute the squared magnitude of a vector."""
        return v[0]**2 + v[1]**2
    
    # Unpack rectangle vertices
    A, B, C, D = rect
    
    # Compute vectors
    AB = vector(A, B)
    AD = vector(A, D)
    AP = vector(A, p)
    
    # Check if the point projects within both sides of the rectangle
    within_AB = 0 <= dot_product(AP, AB) <= magnitude_squared(AB)
    within_AD = 0 <= dot_product(AP, AD) <= magnitude_squared(AD)
    
    return within_AB and within_AD

# Example usage
rectangle = [(620, 200), (1116, 618), (880, 894), (384, 472)]  # Rectangle vertices
```

```python
point = (1106, 639)  # Test point



image = np.zeros((1200,1600,3), np.uint8)
cv.line(image, (620, 200), (1116, 618), [0,0,255], 5) 
cv.line(image, (1116, 618), (880, 894), [0,0,255], 5) 
cv.line(image, (880, 894), (384, 472), [0,0,255], 5) 
cv.line(image, (384, 472), (620, 200), [0,0,255], 5) 
cv.circle(image, point, radius=5, color=(0, 0, 255), thickness=-1)

plt.imshow(image)
plt.show()

print(is_point_in_rectangle(point, rectangle))
```
